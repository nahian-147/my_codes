```python
from Bio import SeqIO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
records_dict = SeqIO.to_dict(SeqIO.parse("GRCh38_latest_genomic.fna", "fasta"))
```

```python
chromoDict = {}
for record in SeqIO.parse("GRCh38_latest_genomic.fna", "fasta"):
    description = record.description
    splittedDescription = description.split(' ')
    if 'chromosome' in splittedDescription:
        name = 'chromosome' + splittedDescription[splittedDescription.index('chromosome') + 1]
    else:
        name = record.description
    if not name in chromoDict:
        chromoDict[name] = []
        chromoDict[name].append(record.seq)
    else:
        chromoDict[name].append(record.seq)
```
