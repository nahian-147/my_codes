## Some Complex Computations ...

```python
z = [] #Empty List
for k in np.linspace(-4,4,100):
    for l in np.linspace(-4,4,100):
        z.append(k+l*1j)

#z is now our complex variable!
z[:3] #Showing first 3 entries
```

```python
C = 3+5j
w = np.power(z,3)-np.power(z,2)+z-C
x = np.real(w)
y = np.imag(w)
plt.figure(figsize=(20,20),dpi=100)
plt.plot(x,y,'.')
```

```python
x = np.linspace(-1,1,10000)
y1 = np.sqrt(1-x*x)/4
y2 = -np.sqrt(1-x*x)/4
z1 = x+y1*1j
z2 = x+y2*1j
z = list(z1)+list(z2)
```

```python
plt.plot(np.real(z),np.imag(z))
```

```python
w = np.power(z,7)-8*np.power(z,5)+5*np.power(z,3)
x = np.real(w)
y = np.imag(w)
plt.figure(figsize=(100,50))
plt.plot(x,y,'.',color='purple')
plt.hlines(0,-2.1,2.1,'k')
plt.vlines(0,-1.2,1.2,'k')
plt.grid()
```
