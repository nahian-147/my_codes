```python
from Bio import SeqIO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ids = []
seqs = []
lns = []
country = []
host = []
year = []
```

```python
records_dict = SeqIO.to_dict(SeqIO.parse("dengue.fa", "fasta"))
records_dict['LC428064']
```

```python
for record in SeqIO.parse("dengue.fa", "fasta"):
    dsc = record.description.split('|')
    ids.append(record.id)
    seqs.append(record.seq)
    lns.append(len(record.seq))
    country.append(dsc[2])
    host.append(dsc[1])
    year.append(dsc[3])
```

```python
records = pd.DataFrame({'id':ids,'seq':seqs,'length':lns,'country':country,'host':host,'year':year})
```

```python
data = records.groupby('country').count()['length']
data
```

```python
plt.figure(figsize=(12,7),dpi=100)
plt.pie(data,labels=data.index,autopct='%.0f%%');
```
