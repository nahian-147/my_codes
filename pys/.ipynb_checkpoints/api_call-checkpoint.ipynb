```python
wind_mph = []
wind_kph = []
wind_degree = []
temp_f = []
temp_c = []
humidity = []
pressure_mb = []
pressure_in = []
precip_mm = []
precip_in = []
lat = []
lon = []
```

```python
weatherDataDictionary = {'lat' : lat,
                         'lon' : lon,
                         'wind_mph' : wind_mph,
                         'wind_kph' : wind_kph,
                         'wind_degree' : wind_degree,
                         'temp_f' : temp_f,
                         'temp_c' : temp_c,
                         'humidity' : humidity,
                         'pressure_mb' : pressure_mb,
                         'pressure_in' : pressure_in,
                         'precip_mm' : precip_mm,
                         'precip_in' : precip_in}
```

```python
lat_max = 26.4
lat_min = 20.3

lon_max = 92.45
lon_min = 88

lats = np.linspace(lat_min,lat_max,10)
lons = np.linspace(lon_min,lon_max,10)
```

```python
def prepareParams(lat,lon):
    pos = str(lat) + ',' + str(lon)
    return {'key' : '62f05a43fda14b868f045838222405','q': pos}
```

```python
queries = []
for lat in lats:
    for lon in lons:
        queries.append(prepareParams(lat,lon))
```

```python
responses = {}
for query in queries:
    response = reqs.get(f"{'https://api.weatherapi.com/v1/current.json'}",params = query).json()
    responses[query['q']] = response
```

```python
for response in responses:
    weatherDataDictionary['lat'].append(responses[response]['location']['lat'])
    weatherDataDictionary['lon'].append(responses[response]['location']['lon'])
    weatherDataDictionary['wind_mph'].append(responses[response]['current']['wind_mph'])
    weatherDataDictionary['wind_kph'].append(responses[response]['current']['wind_kph'])
    weatherDataDictionary['wind_degree'].append(responses[response]['current']['wind_degree'])
    weatherDataDictionary['temp_f'].append(responses[response]['current']['temp_f'])
    weatherDataDictionary['temp_c'].append(responses[response]['current']['temp_c'])
    weatherDataDictionary['humidity'].append(responses[response]['current']['humidity'])
    weatherDataDictionary['pressure_mb'].append(responses[response]['current']['pressure_mb'])
    weatherDataDictionary['pressure_in'].append(responses[response]['current']['pressure_in'])
    weatherDataDictionary['precip_mm'].append(responses[response]['current']['precip_mm'])
    weatherDataDictionary['precip_in'].append(responses[response]['current']['precip_in'])
```

```python
weatherDataFrame = pd.DataFrame(weatherDataDictionary)
weatherDataFrame
```

```python
weatherDataFrame.to_csv('dataFrame.csv')
```
