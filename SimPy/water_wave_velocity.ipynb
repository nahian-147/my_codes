```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l1 = np.linspace(.001,.1,1000)
g = 9.81
rho = 1000
h = 50
s = 7.9*10**-2
```

```python
c1 = ((g*l1/(2*np.pi))*(1+s*4*np.pi**2/(rho*g*l1**2))*np.tanh(2*np.pi*h/l1))**.5
```

```python
plt.figure(figsize=(10,5),dpi=250)
plt.plot(l1,c1)
plt.xlabel('Wavelength labda in m')
plt.ylabel('Speed of Wave in m/s')
```

```python
l2 = np.linspace(1,2000,1000)
c2 = ((g*l2/(2*np.pi))*(1+s*4*np.pi**2/(rho*g*l2**2))*np.tanh(2*np.pi*h/l2))**.5
```

```python
plt.figure(figsize=(10,5),dpi=250)
plt.plot(l2,c2)
plt.xlabel('Wavelength labda in m')
plt.ylabel('Speed of Wave in m/s')
```
