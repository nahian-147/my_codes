```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 2.414*10**-5
B = 247.8
C = 140
celsious = np.linspace(0,100,1000)
T = np.linspace(273,373,1000)
mu_T = A*10**(B/(T-C))
```

```python
plt.figure(figsize=(10,5),dpi=250)
plt.plot(celsious,mu_T)
plt.xlabel('Temperature (C)')
plt.ylabel('Viscosity (Pa s)')
```
