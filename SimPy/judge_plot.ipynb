```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,20)
y = x*(2-x)
plt.figure(figsize=(10,5),dpi=250)
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=250)
y = np.cos(18*np.pi*x)
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=250)
x_1000 = np.linspace(0,2,1000)
y_1000 = np.cos(18*np.pi*x_1000)
plt.plot(x_1000,y_1000)
```
