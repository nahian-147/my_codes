```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
t = 0
x = np.linspace(-4,4,1000)
```

```python
y = np.exp(-np.power(x-3*t,2))*np.sin(3*np.pi*(x-t))
```

```python
plt.figure(figsize=(10,5),dpi=250)
plt.plot(x,y)
```

```python
x, y, t = symbols('x y t')
y = exp(-(x-3*t)**2)*sin(3*pi*(x-t))
simplify(y)
```
